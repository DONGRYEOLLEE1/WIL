# 결정 트리(Decision Tree)
    * 분류와 회귀에 사용되는 지도 학습
    * 데이터 특성으로부터 추론된 결정 규칙을 통해 값을 예측
    * if-then-else 결정 규칙을 통해 데이터학습
    * 트리의 깊이가 깊을수록 복잡한 모델
    * 장점
        * 이해와 해석이 쉬움
        * 시각화 용이
        * 많은 데이터 전처리가 필요하지않음
        * 수치형과 범주형 데이터 모두 다룰수 있음

## 분류 - DecisionTreeClassifier()
        - DecisionTreeClassifier는 분류를 위한 결정트리 모델
        - 두개의 배열 X, y를 입력 받음
            * X는 [n_samples, n_features] 크기의 데이터 특성 배열
            * y는 [n_samples] 크기의 정답 배열

```python
X = [[0, 0], [1, 1]]
y = [0, 1]

model = tree.DecisionTreeClassifier().fit(X, y)
```

```python
model.predict([[2., 2.]])
```

#### 학습된 결정 트리 시각화

```python
model = DecisionTreeClassifier().fit(iris.data, iris.target)
```

```python
# 텍스트를 통한 시각화
r = tree.export_text(decision_tree = model,
                    feature_names=iris.feature_names)
print(r)
```

```python
# plot_tree를 사용한 시각화
tree.plot_tree(model);
```

```python
# graphviz를 사용한 시각화
dot_data = tree.export_graphviz(decision_tree = model,
                               feature_names = iris.feature_names,
                               class_names = iris.target_names,
                               filled = True, rounded = True,
                               special_characters = True)
graph = graphviz.Source(dot_data)
graph
```

```python
# 시각화
n_classes = 3
plot_colors = 'ryb'
plot_step = 0.02
```

```python
from matplotlib import cm
```

```python
# 결정경계 시각화
plt.figure(figsize = (16, 8))

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3], [1, 2], [1,3], [2,3]]):
    X = iris.data[:, pair]
    y = iris.target
    
    model = DecisionTreeClassifier().fit(X, y)
    
    plt.subplot(2, 3, pairidx + 1)
    
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                        np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad = -0.5, w_pad = 0.5, pad = 2.5)
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contour(xx, yy, Z, cmap = cm.RdYlBu)
    
    plt.xlabel(iris.feature_names[pair[0]])
    plt.xlabel(iris.feature_names[pair[1]])
    
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c = color, label = iris.target_names[i],
                   cmap = plt.cm.RdYlBu, edgecolors = 'b', s = 15)
        
plt.suptitle('DT')
plt.legend(loc = 'lower right', borderpad = 0, handletextpad = 0)
plt.axis('tight')
```

```python
# 하이퍼파라미터를 변경해 보면서 결정경계의 변화 확인

plt.figure(figsize = (16, 8))

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3], [1, 2], [1,3], [2,3]]):
    X = iris.data[:, pair]
    y = iris.target
    
    model = DecisionTreeClassifier(max_depth = 2).fit(X, y)
    
    plt.subplot(2, 3, pairidx + 1)
    
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                        np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad = -0.5, w_pad = 0.5, pad = 2.5)
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contour(xx, yy, Z, cmap = cm.RdYlBu)
    
    plt.xlabel(iris.feature_names[pair[0]])
    plt.xlabel(iris.feature_names[pair[1]])
    
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c = color, label = iris.target_names[i],
                   cmap = plt.cm.RdYlBu, edgecolors = 'b', s = 15)
        
plt.suptitle('DT')
plt.legend(loc = 'lower right', borderpad = 0, handletextpad = 0)
plt.axis('tight')
```
