```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
xy = pd.read_csv('Data/diabetes.csv')
xy[:5]
```

```python
x_data = xy[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction',
             'Age']]
y_data = xy['Outcome']
```

```python
print(x_data.shape, y_data.shape)
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim = x_data.shape[1], activation = 'sigmoid'))
tf.model.compile(loss = 'binary_crossentropy', optimizer = tf.keras.optimizers.SGD(learning_rate = .1), metrics = ['acc'])
tf.model.summary()
```

```python
history = tf.model.fit(x_data, y_data, epochs = 500)
```

```python
# acc
print('Accuracy : {}'.format(history.history['acc'][-1]))
```

```python
# predict a single data point
y_predict = tf.model.predict([[0.176471, 0.155779, 0, 0, 0, 0.052161, -0.952178, -0.733333]])
print("Predict: {0}".format(y_predict))
```

```python
# evaluating model
evaluate = tf.model.evaluate(x_data, y_data)
print('loss: {}, accuracy: {}'.format(evaluate[0], evaluate[1]))
```
