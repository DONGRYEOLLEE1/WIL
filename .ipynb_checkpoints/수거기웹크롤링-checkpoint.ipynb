```python

import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://superbin.co.kr/new/contents/location_list.php?skey=&sval=&pg=1'
source = requests.get(url)
soup = BeautifulSoup(source.content, 'html.parser')

txt = soup.select('strong.lc_point')

address = []

for i in range(0, 10):
    add = txt[i].text
    
    address.append(add)

final = pd.DataFrame({'순환자원 회수로봇 설치위치' : address})
final
```
